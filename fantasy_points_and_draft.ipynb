# Converting to Fantasy Points

```python
import pandas as pd
```

```python
player_proj = pd.read_csv('player_proj_df.csv')
player_info = pd.read_csv('player_info_df.csv')
```

```python
df_proj_stats = player_proj.loc[:, 
               ['proj_pts','proj_min','proj_fgm','proj_fga','proj_fg3m','proj_fg3a','proj_ftm','proj_fta','proj_oreb','proj_dreb','proj_ast','proj_stl','proj_tov', 'proj_blk']]

```

```python
df_proj_stats = pd.concat([player_info, df_proj_stats], axis=1)
```

```python
df_proj_stats.rename(columns = {'proj_season_id':'season_id',
                               'proj_pts': 'pts',
                                'proj_min':'min',
                                'proj_fgm':'fgm',
                                'proj_fga':'fga',
                                'proj_fg3m':'fg3m',
                                'proj_fg3a':'fg3a',
                                'proj_ftm':'ftm',
                                'proj_fta':'fta',
                                'proj_oreb':'oreb',
                                'proj_dreb':'dreb',
                                'proj_ast':'ast',
                                'proj_stl':'stl',
                                'proj_tov':'tov',
                               'proj_blk':'blk'}, inplace = True)
```

```python
df_proj_stats.sample(5)
```

```python
def fantasy_pts_converter(row):
    pts = row['pts']
    fgm = row['fgm']
    fga = row['fga']
    ftm = row['ftm']
    fta = row['fta']
    oreb = row['oreb']
    dreb = row['dreb']
    ast = row['ast']
    stl = row['stl']
    tov = row['tov']
    blk = row['blk'] 
    fantasy_pts = pts + fgm - fga + ftm - fta + oreb + dreb + ast + stl - tov + blk
    return fantasy_pts
```

```python
fantasy_value = df_proj_stats.apply(fantasy_pts_converter, axis = 1)
```

```python
df_proj_stats['fantasy_pts'] = fantasy_value
```

```python
df_proj_stats.sort_values('fantasy_pts', ascending = False)
```

```python
df_for_export = round(df_proj_stats.sort_values('fantasy_pts', ascending = False),2)
```

```python
df_for_export.to_csv('projected_fantasy_pts_final_2018_19.csv', index=False)
```

```python
position_dictionary = {
    'PG': 21,
    'SG': 13,
    'SF': 12,
    'PF': 19,
    'C': 15
}
```

```python
df1 = pd.read_csv('Fantasy Projections - 2018_19 - 2018_espn_player_positions.csv', header=0)
df2 = pd.read_csv('projected_fantasy_pts_final_2018_19.csv', header=0)
```

```python
df1.sample()
```

```python
df2.sample()
```

```python
df3 = df1.merge(df2, how='left', on='player_id')
```

```python
df_final = df3[['player_name_x', 'player_id', 'season_id_x', 'espn_position', 'fantasy_pts']]
```

```python
df_final.rename(columns={
    'player_name_x': 'player_name',
    'season_id_x': 'season_id'
}).head(5)
```

```python
df_final.to_csv('projected_fanatasy_pts_with_positions_2018_19.csv', index=False)
```

```python
positions = [
    'PG',
    'SG',
    'SF',
    'PF',
    'C'
]
```

```python
def calc_baseline(position, df):
    df_filter = df[df['espn_position'] == position]
#     print(df_filter.tail(5))
    x = position_dictionary.get(position)
#     print(x)
    replacement_player = df_filter.iloc[x]
#     print(replacement_player)
    value_to_subtract = replacement_player['fantasy_pts']
    return value_to_subtract
```

```python
baseline_points = {}
for position in positions:
        y = calc_baseline(position, df_final)
        baseline_points[position] = y
```
